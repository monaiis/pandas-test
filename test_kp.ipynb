```python
#dfs['Unnamed: 25'] = dfs['Unnamed: 25'].fillna(0)

dict = {}
#use enumerate to keep track of index
for i,e in enumerate(dfs['Unnamed: 2'][2:]):
    if dict.get(e) is None:
        dict[e] = 0
    dict[e] += dfs['Unnamed: 25'][i+2]
    #print(dfs['Unnamed: 25'][i+2])
```

```python
#getting key of maximum value in dictionary <--- Ans1

max(dict, key=dict.get)
```

```python
#read sales_revenue.xlsx in US-YTD worksheet
#I assume that US-YTD is the summary of Monthly-States so I will pick this to answer objective 2
df = pd.read_excel ('sales_revenue.xlsx', 'US-YTD')
df
```

```python
#filter only '.' in month column
df2 = df.loc[df['Unnamed: 1'] == '.']
df2
```

```python
#select sales column of each categories
RES = df2['Unnamed: 4'].sum()
COM = df2['Unnamed: 8'].sum()
IN = df2['Unnamed: 12'].sum()
#cast '.' to NaN
temp = pd.to_numeric(df2['Unnamed: 16'], errors='coerce')
TRAN = temp.sum()
```

```python
#print to compare the results, you will see that RESIDENTIAL category use elec the most <--- Ans2
print(f"RESIDENTIAL = {RES} \nCOMMERCIAL = {COM} \nINDUSTRIAL = {IN} \nTRANSPORTATION = {TRAN}")
```
